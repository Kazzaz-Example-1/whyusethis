```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect("northwind-SQLite3/Northwind_small.sqlite")
```

```python
queryTables = """
select name from sqlite_master
where type = 'table'
"""

dfTables = pd.read_sql_query(queryTables, con)
```

```python
print(dfTables)
```

```python
queryEmployee = """
select * from employee
"""

dfEmployee = pd.read_sql_query(queryEmployee, con)

dfEmployee.head(10)
```

```python
for col in dfEmployee:
    print(col)
```

```python
queryEmployeeAge = """
select LastName, FirstName, BirthDate, 
date('now') - BirthDate as age from Employee"
"""

dfEmployeeAge = pd.read_sql_query(queryEmployeeAge, con)

print(dfEmployeeAge)
```

```python
dfEmployeeAge["age"].mean()
```
